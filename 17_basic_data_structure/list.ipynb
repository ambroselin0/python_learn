# List

A lots of data structure can be made with python list, but not all coding language have this feature.

## disordered list

A list with no order, it is just a collection of items where each item holds a relative position with respect to the others.

### Python Comprehension

- init() - create a new list that is empty
- add(item) - add a new item to the list
- remove(item) - remove the item from the list
- search(item) - search for the item in the list
- is_empty() - check if the list is empty
- length() - return the number of items in the list
- append(item) - add a new item to the end of the list
- index(item) - return the position of item in the list
- insert(pos, item) - add a new item to the position of the list
- pop() - remove and return the last item in the list
- pop(pos) - remove and return the item in the position of the list

## ordered list

A list with order.

### Python Comprehension

- init(order) - create a new list that is empty
- add(item) - add a new item to the list
- remove(item) - remove the item from the list
- search(item) - search for the item in the list
- is_empty() - check if the list is empty
- length() - return the number of items in the list
- index(item) - return the position of item in the list
- pop() - remove and return the last item in the list
- pop(pos) - remove and return the item in the position of the list

```python
class Node(object):
    def __init__(self, data):
        self.data = data
        self.next = None

    def get_data(self):
        return self.data

    def get_next(self):
        return self.next

    def set_data(self, data):
        self.data = data

    def set_next(self, next_node):
        self.next = next_node


class DisorderedList(object):
    def __init__(self):
        self.head = None

    def is_empty(self):
        return self.head is None

    def add(self, item):
        node = Node(data=item)
        node.set_next(self.head)
        self.head = node

    def __len__(self):
        count = 0
        current: Node = self.head
        while current is not None:
            count += 1
            current = current.get_next()
        return count

    def search(self, item):
        current: Node = self.head
        while current is not None:
            value = current.get_data()
            if value == item:
                return True
            current = current.get_next()
        return False

    def remove(self, item):
        current: Node = self.head
        previous = None
        found = False
        while not found:
            if current.get_data() == item:
                found = True
            else:
                previous = current
                current = current.get_next()

        if previous is None:
            self.head = current.get_next()
        else:
            previous.set_next(current.get_next())

        del current

    def append(self, item):
        current: Node = self.head
        while current.get_next() is not None:
            current = current.get_next()
        node = Node(item)
        current.set_next(node)

    def index(self, item):
        i = 0
        current = self.head
        while current is not None:
            if current.get_data() == item:
                return i
            i += 1
            current = current.get_next()

    def insert(self, pos, item):
        if pos == 0:
            self.add(item)
            return None
        previous = None
        current = self.head
        node = Node(item)
        for j in range(pos):
            previous: Node = current
            current = current.get_next()
        previous.set_next(node)
        node.set_next(current)

    def pop(self, pos=None):
        if pos is None:
            previous = None
            current:Node = self.head
            while current.get_next() is not None:
                previous:Node = current
                current = current.get_next()
            previous.set_next(None)
            current_value = current.get_data()
            del current
            return current_value
        else:
            previous = None
            current:Node = self.head
            for j in range(pos):
                previous: Node = current
                current = current.get_next()
            if previous is None:
                self.head = current.get_next()
            else:
                previous.set_next(current.get_next())

            current_value = current.get_data()
            del current
            return current_value


a = DisorderedList()
for i in range(5):
    a.add(i)
print(len(a))
print(a.search(3))
print(a.index(1))
a.insert(2, 1) # 4 3 1 2 1 0
print(a.pop()) # 4 3 1 2 1
a.pop(0) # 3 1 2 1
a.pop(1)

```
